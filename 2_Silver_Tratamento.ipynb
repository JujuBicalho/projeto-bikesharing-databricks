```python
############# CAMADA SILVER - TRATAMENTO #############
spark.sql("USE projeto_bikes_enterprise")

from pyspark.sql.functions import col, to_date, when, round, lit

df_bronze = spark.read.table("bronze_bikes_day")

display(df_bronze)
```

```python
from pyspark.sql.functions import col, to_date, when, round, lit

########### LIMPEZA/TRATAMENTO ###########

df_silver = df_bronze.select(
    col("instant").alias("id"),
    to_date(col("dteday"), "yyyy-MM-dd").alias("data"),
    when(col("yr") == 0, 2011).otherwise(2012).alias("ano"),
    col("mnth").alias("mes"),
    when(col("holiday") == 1, "Sim").otherwise("No").alias("feriado"),
    when(col("weekday") == 0, "Domingo")
    .when(col("weekday") == 1, "Segunda")
    .when(col("weekday") == 2, "Tera")
    .when(col("weekday") == 3, "Quarta")
    .when(col("weekday") == 4, "Quinta")
    .when(col("weekday") == 5, "Sexta")
    .when(col("weekday") == 6, "Sbado")
         .alias("dia_semana"),
    when(col("workingday") == 1, "Dia til").otherwise("Fim de semana").alias("dia_util"),
    col("cnt").alias("total_aluguel"),
    
    # --- BLOCO 1: ESTAO ---
    when(col("season") == 1, "Inverno")
    .when(col("season") == 2, "Primavera")
    .when(col("season") == 3, "Vero")
    .otherwise("4. Outono").alias("estacao"),  

    # --- BLOCO 2: CLIMA  ---
    when(col("weathersit") == 1, "Limpo/Parcial")
    .when(col("weathersit") == 2, "Nublado")
    .when(col("weathersit") == 3, "Chuva Leve/Neve")
    .otherwise("Chuva Forte").alias("clima"), 
    
    # --- BLOCO 3: MATEMTICA ---
    round(col("temp") * 41, 1).alias("temperatura"),        
    round(col("atemp") * 50, 1).alias("sensacao_termica"), 
    round(col("hum") * 100, 1).alias("umidade"),
    round(col("windspeed") * 67, 1).alias("velocidade_vento") 
    
)

display(df_silver)
df_silver.write.mode("overwrite").saveAsTable("silver_bikes_tratadas")
```
