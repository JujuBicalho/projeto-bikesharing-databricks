```python
#Configuração do ambiente
#Criar um banco de dados isolado

spark.sql("CREATE DATABASE IF NOT EXISTS projeto_bikes_enterprise")
spark.sql("USE projeto_bikes_enterprise")

print("Configuração concluída do banco de dados 'projeto_bikes_enterprise'")
```

```python
############# CAMADA BRONZE - INGESTÃO #############
path_day = "/databricks-datasets/bikeSharing/data-001/day.csv"
path_hour = "/databricks-datasets/bikeSharing/data-001/hour.csv"

df_raw_day = spark.read.csv(path_day, header=True, inferSchema=True)
df_raw_hour = spark.read.csv(path_hour, header=True, inferSchema=True)

#Salvar como tabela
#-Tabela de dias
try:
    df_raw_day.write.mode("overwrite").saveAsTable("bronze_bikes_day")
    print("Tabela [bronze_bikes_dia] salva com sucesso!")
except Exception as erro:
    print(f"Erro ao salvar tabela [bronze_bikes_dia]. Motivo: {erro}")
    raise erro  #Faz o Databricks a marcar o Job como "Failed" e a camada Silver não rodará.

#-Tabela de horas
try:
    df_raw_hour.write.mode("overwrite").saveAsTable("bronze_bikes_hour")
    print("Tabela [bronze_bikes_hour] salva com sucesso!")
except Exception as erro:
    print(f"Erro ao salvar tabela [bronze_bikes_hour]. Motivo: {erro}")
    raise erro 
```
